```python
import pandas as pd
import sklearn
import numpy as np
```

```python
M = np.genfromtxt('../data/processed.va.data', missing_values=0, delimiter=',', dtype=float)
va_df = pd.DataFrame(M,columns=["age","sex","cp","trestbps","chol","fbs","restecg","thalach","exang","oldpeak","slope","ca","thal","num"],)
print(va_df)
```

```python
va_df['ca'].fillna(va_df['ca'].mode()[0], inplace = True)
va_df['thal'].fillna(va_df['thal'].mode()[0], inplace = True)
va_df.isnull().sum()
```

```python
del va_df['ca']
del va_df['thal']
```

```python
va_df.dropna(axis=0,inplace=True,subset=['trestbps','thalach','exang','oldpeak'])
va_df.isnull().sum()
```

```python
va_df['chol'].fillna(va_df['chol'].mean(), inplace = True)
va_df['fbs'].fillna(va_df['fbs'].mode()[0], inplace = True)
va_df['slope'].fillna(va_df['slope'].mode()[0], inplace = True)
va_df.isnull().sum()
```

```python
va_df['heartdis'] = np.where(va_df['num'] > 0.0, 1,0)
del va_df['num']
```

```python
va_df.to_csv("../data/ProcessedVAData.csv",index=False)
va_df
```
