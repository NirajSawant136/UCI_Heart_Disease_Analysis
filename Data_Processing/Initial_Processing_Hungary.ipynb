```python
import pandas as pd
import sklearn
import numpy as np
```

```python
M = np.genfromtxt('../data/reprocessed.hungarian.data', missing_values=0, delimiter=' ', dtype=float)
hungary_df = pd.DataFrame(M,columns=["age","sex","cp","trestbps","chol","fbs","restecg","thalach","exang","oldpeak","slope","ca","thal","num"],)
print(hungary_df)
```

```python
del hungary_df['ca']
del hungary_df['thal']
```

```python
hungary_df=hungary_df.replace(-9.0,np.NaN)
hungary_df.isnull().sum()
```

```python
hungary_df['trestbps'].fillna(hungary_df['trestbps'].mean(), inplace = True)
hungary_df['chol'].fillna(hungary_df['chol'].mean(), inplace = True)
hungary_df['fbs'].fillna(hungary_df['fbs'].mode()[0], inplace = True)
hungary_df['restecg'].fillna(hungary_df['restecg'].mode()[0], inplace = True)
hungary_df['thalach'].fillna(hungary_df['thalach'].mode()[0], inplace = True)
hungary_df['exang'].fillna(hungary_df['exang'].mode()[0], inplace = True)
hungary_df['slope'].fillna(hungary_df['slope'].mode()[0], inplace = True)
hungary_df.isnull().sum()
```

```python
hungary_df['heartdis'] = np.where(hungary_df['num'] > 0.0, 1,0)
del hungary_df['num']
```

```python
hungary_df.to_csv("../data/ProcessedHungaryData.csv",index=False)
hungary_df
```
