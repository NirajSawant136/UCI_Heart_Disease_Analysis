```python
import pandas as pd
import sklearn
import numpy as np
```

```python
M = np.genfromtxt('../data/processed.switzerland.data', missing_values=0, delimiter=',', dtype=float)
swiss_df = pd.DataFrame(M,columns=["age","sex","cp","trestbps","chol","fbs","restecg","thalach","exang","oldpeak","slope","ca","thal","num"],)
print(swiss_df)
```

```python
del swiss_df['ca']
del swiss_df['thal']
```

```python
swiss_df['trestbps'].fillna(swiss_df['trestbps'].mean(), inplace = True)
swiss_df['oldpeak'].fillna(swiss_df['oldpeak'].mean(), inplace = True)
swiss_df['fbs'].fillna(swiss_df['fbs'].mode()[0], inplace = True)
swiss_df['restecg'].fillna(swiss_df['restecg'].mode()[0], inplace = True)
swiss_df['thalach'].fillna(swiss_df['thalach'].mode()[0], inplace = True)
swiss_df['exang'].fillna(swiss_df['exang'].mode()[0], inplace = True)
swiss_df['slope'].fillna(swiss_df['slope'].mode()[0], inplace = True)
swiss_df.isnull().sum()
```

```python
swiss_df['heartdis'] = np.where(swiss_df['num'] > 0.0, 1,0)
del swiss_df['num']
```

```python
swiss_df.to_csv("../data/ProcessedSwissData.csv",index=False)
swiss_df
```
