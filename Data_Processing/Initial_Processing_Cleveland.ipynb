```python
import pandas as pd
import sklearn
import numpy as np
```

```python
M = np.genfromtxt('../data/processed.cleveland.data', missing_values=0, delimiter=',', dtype=float)
cleveland_df = pd.DataFrame(M,columns=["age","sex","cp","trestbps","chol","fbs","restecg","thalach","exang","oldpeak","slope","ca","thal","num"],)
print(cleveland_df)
```
